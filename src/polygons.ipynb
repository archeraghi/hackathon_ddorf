```python
import requests
import json
import pandas as pd
bezirk_geojson = requests.get('https://opendata.duesseldorf.de/sites/default/files/Stadtbezirke%202017%20Zuordnung%20Stadtteile%20-%20WGS84.geojson').json()
```

```python
ids = [('0' + str(id))[-2:] for id in range(1,11)]
```

```python
def swapcoords(bezirk):
    return [
        [bezirk_geojson['features'][0]['geometry']['coordinates'][0][1],bezirk_geojson['features'][0]['geometry']['coordinates'][0][0]]
        for coords in 
        bezirk_geojson['features'][0]['geometry']['coordinates'][0]
    ]
    
```

```python
path = '../data/Stadtbezirk_'+'01'+'.json'
with open(path) as f:
    bezirk = json.load(f)
```

```python
coords = bezirk['features'][0]['geometry']['coordinates'][0]
```

```python
[
    {'lat':pair[1],'lon':pair[0]}
    for pair
    in coords
]
```

```python
for id in ids:
    path = '../data/Stadtbezirk_'+id+'.json'
    bezirk = json.load(path)['features']
    bezirk[0]['geometry']['coordinates'][0] = swapcoords(bezirk)
    with open(path,'w') as f:
        json.dump(bezirk,f)
```
